# Panel Response Notes

### Central Learnings

- Powershell script to determine if old soldiers had their data in systems
- Got rid of old data so that nobody can find and use the old data for malicious reasons
- Employee on leave of absence had their access shut down because they had stolen from the company and interacted with competetors
- Cyber security allowed this threat to be stopped, and legal actions were taken
- Biggest cybersecurity threats: Ransomware --> can kill companies, costs people many billions of dollars
- Ransomware can steal ideas that people invested a lot in
- AI contributes to ransonware and more advanced cyber attack weapons
- Fishing: Users who don't know security basics click on links, hackers can access classified data
- Zero Day = Vulnerability that has no patch for it, **everyone** is vulnerable --> ex. malware can apply to all updates from a cloud 
- Use real world threat examples to help prepare for actual cyber threats
- Backdoor = Unconventional way into a software system (can help hackers get in) --> Leads to ethical concerns
- Penetration testers are used in order to ensure security of systems
- Software is a career that was not prominent a few decades ago, but has now evolved into a crutial topic in the modern world
- Getting a background in I.T helps with a cyber security career path
- Soft skills very essential to a software career (or any career in general) --> Curious, passionate, likable
- Learn how networks communicate, internships to learn the field in the real world
- From a business standpoint, cyber security is important to investors not to obtain more money, but to prevent themselves from losing money
- Cyber security is great in itself, but the people behind it are just as important
- Trend for next 5-10 years: Reduction of human error through specialization (ex. medical field), AI, prioritization of cyber security
- Networking is extremely important with a career in software
- Computer scientists must adapt their competency because cyber security is a constantly updating and changing field
- Could use TACP server to experiment with cyber security for my website
- Could also use CTF environments for cyber security 
- CC certification would be a good certification that is free and allows for learning
- Most technical exploits and cyber attacks are never actually implemented
- Cyber attacks can spread extremely quickly, and even faster for legal cyber processes --> "Patch Tuesday, exploit Wednesday"
- Meeting and speaking with diverse groups of people can help to keep updated on software advances

### Key Takeaways:

- Cybersecurity threats are rapidly evolving, requiring constant learning and adaptation  
- Real-world case studies help reinforce awareness and preparedness  
- Both technical and soft skills are essential in the cybersecurity field  
- AI is both a tool and a threat in modern cyber defense and attacks  
- Investing in cybersecurity is crucial for protecting company value and reputation  
- Networking, internships, and certifications are key stepping stones into the field
